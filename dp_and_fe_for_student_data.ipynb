```python
print("ก่อนลบ")
print(df)
```

dropna จะลบแถวทั้งหมดออกจาก df ที่มีค่า NaN

```python
#วิธีที่ 2 เติมค่าด้วยค่าเฉลี่ย
mean_age = df['อายุ'].mean()
df['อายุ'].fillna(mean_age, inplace=True)

print(f"\nค่าเฉลี่ยอายุ - {mean_age:.1f} ปี")
print("หลังเติมค่า")
print(df)
```

คำนวณค่าเฉลี่ยของข้อมูลในคอลัมน์ 'อายุ' แล้วเก็บไว้ในตัวแปร mean_age แล้วค้นหาค่า NaN ทั้งหมดในคอลัมน์ 'อายุ' แล้วแทนที่ด้วยค่า mean_age ที่คำนวณได้ โดยที่ inplace=True หมายถึงการแก้ไขข้อมูลใน df โดยตรง

```python
#วิธีที่ 3 เติมค่าด้วยค่าที่พบบ่อยที่สุด
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)

print("\nหลังเติมค่า")
print(df)
```

เติมค่าว่างในคอลัมน์ 'คณะ' ด้วยค่าที่ปรากฏบ่อยที่สุด ของคอลัมน์นั้น โดยแก้ไขใน df โดยตรง

##การทำให้ข้อมูลเป็นมาตรฐาน

```python
# ตัวอย่าง - ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวกรรม', 'วิศวะ','Engineering','วิศวฯ']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name

# ใช้งาน
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

ใช้ฟังก์ชัน standardize_faculty เพื่อรวมชื่อคณะที่สะกดต่างกันให้เป็นมาตรฐานเดียวกันแล้วนำไปใช้สร้างคอลัมน์ใหม่ชื่อ คณะ_มาตรฐาน ใน df

##การจัดการค่าผิดปกติ

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")
```

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1
```

คำนวณหาค่าควอร์ไทล์ที่ 1 และ ค่าควอร์ไทล์ที่ 3 ของข้อมูลในคอลัมน์ 'อายุ' แล้วหาผลต่างระหว่าง Q3 และ Q1 แล้วเก็บไว้ในตัวแปร IQR

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

คำนวณหาขอบเขตล่างและขอบเขตบนเพื่อใช้ระบุค่าผิดปกติ
lower_bound จะคำนวณค่าขอบเขตล่าง โดยนำ Q1 มาลบด้วย 1.5 เท่าของ IQR
ีupper_bound จะคำนวณค่าขอบเขตบน โดยนำ Q3 มาบวกด้วย 1.5 เท่าของ IQR

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

ทำให้ df ไม่มีแถวซึ่งคอลัมน์ 'อายุ' มีค่าอยู่นอกช่วง lower_bound และ upper_bound ซึ่งใช้ระบุค่าผิดปกติ

```python
print(f"\nขอบเขตล่าง - {lower_bound}")
print(f"ขอบเขตบน - {upper_bound}")
print(f"\nข้อมูลหลังกรองค่าผิดปกติ")
print(df_no_outliers)
print(f"\nค่าเฉลี่ยอายุ (หลังกรองค่าผิดปกติ) - {df_no_outliers['อายุ'].mean():.2f}")
```

##การสร้าง Feature จากวันเวลา

```python
df['วันที่ลงทะเบียน'] = pd.to_datetime(df['วันที่ลงทะเบียน'])

df['เดือนที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.month
df['ปีที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.year
df['วันในสัปดาห์ที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.day_name()
df['วันในเดือนที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.day
```

แปลงคอลัมน์ 'วันที่ลงทะเบียน' ให้เป็นรูปแบบวันที่-เวลา และสร้างคอลัมน์ใหม่ 4 คอลัมน์ ได้แก่ 'เดือนที่ลงทะเบียน', 'ปีที่ลงทะเบียน', 'วันในสัปดาห์ที่ลงทะเบียน' และ 'วันในเดือนที่ลงทะเบียน' เพื่อแยกข้อมูล

```python
print(df[['วันที่ลงทะเบียน', 'เดือนที่ลงทะเบียน', 'ปีที่ลงทะเบียน', 'วันในสัปดาห์ที่ลงทะเบียน']].head())
```

##การรวม Features

```python
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm']/100)**2
df['สุขภาพ'] = pd.cut(df['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```

โค้ดนี้คำนวณค่า BMI โดยการนำค่าน้ำหนัก มาหารด้วยส่วนสูง ที่แปลงเป็นเมตร ยกกำลังสอง ตามสูตร BMI  และ จัดหมวดหมู่สุขภาพตาม BMI ออกเป็น 4 ช่วง เป็น 'ผอม', 'ปกติ', 'อ้วน', และ 'อ้วนมาก'
